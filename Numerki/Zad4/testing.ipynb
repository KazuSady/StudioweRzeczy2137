```python
import numpy as np

# Funkcje
def f1(x):
    return x * x + 3
def f2(x):
    return np.cos(2 * x)
def f3(x):
    return 4 * np.log(x + 3)
#------------------------------------------------------
# Horner
def Horner(stopien, a, x):
    wynik = a[stopien]
    for i in range(stopien - 1, -1, -1):
        wynik = wynik * x + a[i]
    return wynik

#------------------------------------------------------
# Simpson
def Simpson(a, b, funkcja, e):
    temp = 0
    przyb = 0
    h = (b - a) / 3
    while True:
        przyb += (b-a)/6 * (funkcja(a) + 4 * funkcja((a+b)/2) + funkcja(b))
        if abs(przyb - temp) < e:
            break
        temp = przyb
        a += h
    return przyb

#------------------------------------------------------
import math

def composite_simpson(function, x, y, eps, step):
    """
    Computes the definite integral of the function on the interval [x, y] with the given
    accuracy using the composite Simpson's rule. The interval is divided into subintervals
    of length 'step' and the accuracy is checked after each iteration.
    """
    result = 0
    helper2 = 0
    k = 0
    
    while True:
        k += 1
        helper2 = result
        j = 1
        
        while True:
            helper = result
            a = x
            b = y
            result = 0
            c = (b - a) / j

            for i in range(j):
                b = a + c
                result += ((b - a) / 6) * (function(a) + 4*function((a + b) / 2) + function(b))
                a += c

            if abs(helper - result) <= eps and j > 2:
                break

            j += 1  # Increase the number of subintervals by a factor of 2
        if x < 0:
            x -= step
        else:
            y += step
        if abs(helper2 - result) < eps:
            break
    return result

# Menu
print("Wybierz funkcję:\n1 - x^2 + 3\n2 - cos(2x)\n3 - 4log(x+3)")
choice = input()
print("Podaj dolną granicę:")
a = float(input())
print("Podaj górną granicę:")
b = float(input())
print("Podaj dokładność:")
e = float(input())

match choice:
    case '1':
        # przyb_cal = Simpson(a, b, f1, e)
        przyb_cal = composite_simpson(f1,a,b,e,0.1)
    case '2':
        # przyb_cal = Simpson(a, b, f2, e)
        przyb_cal = composite_simpson(f2,a,b,e,0.1)
    case '3':
        # przyb_cal = Simpson(a, b, f3, e)
        przyb_cal = composite_simpson(f3,a,b,e,0.1)
print(przyb_cal)
 
#------------------------------------------------------   
      
```
